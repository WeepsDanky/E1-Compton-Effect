```python
import pandas as pd
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define Gaussian function
def gaussian(x, a, x0, sigma):
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))

# Function to fit Gaussian and identify peak
def fit_gaussian(file_path, source_name):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path, delimiter=',')
    
    # Convert columns to numeric
    df = df.apply(pd.to_numeric, errors='coerce')
    
    # Drop rows with all NaN values
    df = df.dropna(how='all')
    
    # Extract relevant columns for analysis
    df_plot = df[['Energy E / keV', 'Events N']].dropna()

    # Apply specific filtering for co60
    if source_name == 'co60':
        df_plot = df_plot[df_plot['Energy E / keV'] > 1000]  # Ignore all peaks before 1000 keV

    # Identify peaks
    peaks, _ = find_peaks(df_plot['Events N'], height=0.1 * max(df_plot['Events N']))

    if len(peaks) > 0:
        x_data = df_plot['Energy E / keV'].values
        y_data = df_plot['Events N'].values

        if source_name == 'co60':
            # Define ranges for the two peaks
            peak_1_range = (x_data > 1130) & (x_data < 1230)  # Around 1173 keV
            peak_2_range = (x_data > 1280) & (x_data < 1420)  # Around 1332 keV

            # Fit Gaussian to the first peak
            x_peak_1 = x_data[peak_1_range]
            y_peak_1 = y_data[peak_1_range]
            initial_guesses_1 = [max(y_peak_1), 1173, 5]  # Initial guesses for Gaussian
            popt1, _ = curve_fit(gaussian, x_peak_1, y_peak_1, p0=initial_guesses_1, maxfev=10000)

            # Fit Gaussian to the second peak
            x_peak_2 = x_data[peak_2_range]
            y_peak_2 = y_data[peak_2_range]
            initial_guesses_2 = [max(y_peak_2), 1332, 5]  # Initial guesses for Gaussian
            popt2, _ = curve_fit(gaussian, x_peak_2, y_peak_2, p0=initial_guesses_2, maxfev=10000)

            # Extract fitted peak positions
            fitted_peaks = [popt1[1], popt2[1]]  # Peak positions (x01, x02)

            # Plot the data with Gaussian fits
            plt.figure()
            plt.plot(x_data, y_data, label=f'{source_name} Spectrum')
            plt.plot(x_peak_1, gaussian(x_peak_1, *popt1), label='Gaussian Fit (Peak 1)', linestyle='--')
            plt.plot(x_peak_2, gaussian(x_peak_2, *popt2), label='Gaussian Fit (Peak 2)', linestyle='--')
            for peak in fitted_peaks:
                plt.axvline(x=peak, color='red', linestyle='--', label=f'Peak at {peak:.2f} keV')
            plt.xlabel('Energy (keV)')
            plt.ylabel('Event Count')
            plt.title(f'{source_name}: Two Independent Gaussian Fits')
            plt.legend()
            plt.grid(True)
            plt.show()

            return fitted_peaks
        else:
            # Fit a single Gaussian for other sources
            peak_index = peaks[np.argmax(df_plot.iloc[peaks]['Events N'].values)]
            peak_energy = df_plot.iloc[peak_index]['Energy E / keV']

            # Narrow range around the peak
            peak_range = (x_data > peak_energy - 10) & (x_data < peak_energy + 10)
            x_peak = x_data[peak_range]
            y_peak = y_data[peak_range]

            initial_guesses = [max(y_peak), peak_energy, 1]  # Initial guesses for Gaussian
            popt, _ = curve_fit(gaussian, x_peak, y_peak, p0=initial_guesses, maxfev=10000)
            
            # Extract fitted peak position
            fitted_peak = popt[1]

            # Plot the data with Gaussian fit
            plt.figure()
            plt.plot(x_data, y_data, label=f'{source_name} Spectrum')
            plt.plot(x_peak, gaussian(x_peak, *popt), label='Gaussian Fit', linestyle='--')
            plt.axvline(x=fitted_peak, color='red', linestyle='--', label=f'Peak at {fitted_peak:.2f} keV')
            plt.xlabel('Energy (keV)')
            plt.ylabel('Event Count')
            plt.title(f'{source_name}: Gaussian Fit')
            plt.legend()
            plt.grid(True)
            plt.show()

            return [fitted_peak]
    else:
        return None

# File paths and source names
file_paths = [
    'data/calibration/am241-sspht002-2048bin-600s-2gain.csv',
    'data/calibration/co57-sspht030-2048bin-600s-2gain.csv',
    'data/calibration/cs137-sspht022-2048bin-600s-2gain.csv', 
    'data/calibration/co60-sspht030-2048bin-600s-2gain.csv'
]
source_names = ['am241', 'co57', 'cs137', 'co60']

# Expected peak values for each source
expected_peaks = {
    'am241': [59.54],
    'co57': [122.06],
    'cs137': [661.66], 
    'co60': [1173.24, 1332.49]  # Two peaks for co60
}

# Extract the observed peaks
observed_peaks = {}
for file_path, source_name in zip(file_paths, source_names):
    peak_energies = fit_gaussian(file_path, source_name)
    if peak_energies is not None:
        observed_peaks[source_name] = peak_energies

# Flatten the observed and expected peaks for linear regression
observed_flat = []
expected_flat = []
for source, peaks in observed_peaks.items():
    observed_flat.extend(peaks)
    expected_flat.extend(expected_peaks[source])

# Create a DataFrame for linear regression
df_peaks = pd.DataFrame({
    'Observed': observed_flat,
    'Expected': expected_flat
})

# Perform linear regression to find the mapping relationship
X = df_peaks[['Observed']]
y = df_peaks['Expected']

model = LinearRegression()
model.fit(X, y)

# Print the linear relationship
slope = model.coef_[0]
intercept = model.intercept_
print(f"Linear relationship: Expected = {slope:.4f} * Observed + {intercept:.4f}")

# Apply the linear relationship to the observed peaks
df_peaks['Calibrated'] = model.predict(X)
print(df_peaks)

# Plot the observed vs. expected peaks and the linear mapping
plt.figure()
plt.scatter(df_peaks['Observed'], df_peaks['Expected'], color='blue', label='Observed vs. Expected')
plt.plot(df_peaks['Observed'], df_peaks['Calibrated'], color='red', label='Calibration Fit')
plt.xlabel('Observed Energy (keV)')
plt.ylabel('Expected Energy (keV)')
plt.title('Calibration Curve')
plt.legend()
plt.grid(True)
plt.show()

# Print the linear mapping formula
print(f"Linear Mapping Formula: Expected Energy (keV) = {slope:.4f} * Observed Energy (keV) + {intercept:.4f}")
```

```python
def calibrate_energy(observed_energy, slope=0.9927267168433642, intercept=-21.694981970559468):
    """
    Calibrate the observed energy readings using the given slope and intercept.

    Parameters:
    - observed_energy (float or array-like): The observed energy readings to be calibrated.
    - slope (float): The slope of the calibration curve.
    - intercept (float): The intercept of the calibration curve.

    Returns:
    - float or array-like: The calibrated energy readings.
    """
    return slope * observed_energy + intercept

# Example usage
observed_energy = 688  # Example observed energy value
calibrated_energy = calibrate_energy(observed_energy)
print(f"Calibrated Energy: {calibrated_energy:.2f} keV")
```

# Expected Compton Scattering Value

```python
import numpy as np

# Constants
M_E_C2 = 511  # keV (electron rest mass energy)

# Dictionary of common photon sources with their primary photon energy in keV
PHOTON_ENERGIES = {
    "cs137": 661.7,  # Cs-137 gamma-ray source
    "co60-2": 1332.5,  # Co-60 gamma-ray source
    "co60-1": 1173.2,  # Co-60 gamma-ray source
    "na22-1": 1274.5,  # Na-22 gamma-ray source
    "na22-2": 511.0,   # Na-22 gamma-ray source
    "co57": 122.0,   # Co-57 gamma-ray source
    "am241": 59.5     # Am-241 gamma-ray source
}

def compton_energy_for_source(source, angle_deg):
    """
    Calculate the expected scattered energy for a given source and angle.
    
    Parameters:
    - source (str): Name of the source (e.g., 'cs137').
    - angle_deg (float): Scattering angle in degrees.
    
    Returns:
    - float: Theoretical scattered photon energy in keV.
    """
    # Get the initial photon energy for the source
    source = source.lower()
    if source not in PHOTON_ENERGIES:
        raise ValueError(f"Source '{source}' not found in the photon energies database.")
    
    E = PHOTON_ENERGIES[source]  # Initial photon energy in keV
    theta_rad = np.radians(angle_deg)  # Convert angle to radians
    
    # Compton formula
    E_prime = E / (1 + (E / M_E_C2) * (1 - np.cos(theta_rad)))
    return E_prime

# Example usage
source_name = "cs137"
scattering_angle = 45  # in degrees
scattered_energy = compton_energy_for_source(source_name, scattering_angle)

print(f"Theoretical scattered energy for {source_name.upper()} at {scattering_angle}°: {scattered_energy:.2f} keV")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import math


# Define the compton_energy_for_source function
def compton_energy_for_source(source, theta_degrees):
    # Define incident energies for different sources
    source_energies = {
        'na22-2': 511.0,  # Na-22 gamma energy in keV
        'co60-1': 1173.2,  # Co-60 gamma energy in keV
        'co60-2': 1332.5,  # Co-60 gamma energy in keV
        # Add more sources if needed
    }
    incident_energy_keV = source_energies.get(source, None)
    if incident_energy_keV is None:
        raise ValueError(f"Unknown source: {source}")
    mec2 = 511.0  # Electron rest mass energy in keV
    theta_radians = math.radians(theta_degrees)
    denominator = 1 + (incident_energy_keV / mec2) * (1 - math.cos(theta_radians))
    scattered_energy = incident_energy_keV / denominator
    return scattered_energy

# Define a smoothing function using a moving average
def smooth_data(data, window_size):
    return data.rolling(window=window_size, center=True).mean()

# Read the CSV files into DataFrames
with_al_data = pd.read_csv('data/experiment/1205/na22-sspht031-2048bin-600s-2gain-45degree-with-al.csv')
no_al_data = pd.read_csv('data/experiment/1205/na22-sspht031-2048bin-600s-2gain-45degree-no-al.csv')

# Ensure the data is aligned by the same 'Energy E / keV' column
with_al_data = with_al_data.sort_values(by='Energy E / keV').reset_index(drop=True)
no_al_data = no_al_data.sort_values(by='Energy E / keV').reset_index(drop=True)

# Smooth the data
window_size = 10  # Adjust the window size as needed
with_al_data['Events N'] = smooth_data(with_al_data['Events N'], window_size)
no_al_data['Events N'] = smooth_data(no_al_data['Events N'], window_size)

# Subtract 'no-al-data' from 'with-al-data'
subtracted_data = with_al_data.copy()
subtracted_data['Events N'] = with_al_data['Events N'] - no_al_data['Events N']

# Identify the peak of the subtracted data
peak_index = subtracted_data['Events N'].idxmax()
peak_energy = subtracted_data.loc[peak_index, 'Energy E / keV']
peak_value = subtracted_data.loc[peak_index, 'Events N']

# Narrow down the data range around the peak
range_width = 100  # Adjust the range width as needed
subtracted_data_peak = subtracted_data[(subtracted_data['Energy E / keV'] >= peak_energy - range_width) &
                                       (subtracted_data['Energy E / keV'] <= peak_energy + range_width)]

# Define a Gaussian function
def gaussian(x, a, x0, sigma):
    return a * np.exp(-(x - x0) ** 2 / (2 * sigma ** 2))

# Fit a Gaussian to the peak
x_data = subtracted_data_peak['Energy E / keV']
y_data = subtracted_data_peak['Events N']
initial_guess = [peak_value, peak_energy, 10]  # Adjust the initial sigma if needed
popt, _ = curve_fit(gaussian, x_data, y_data, p0=initial_guess, maxfev=10000)

# Extract the fitted parameters
a, x0, sigma = popt
fitted_peak_value = a
fitted_peak_energy = x0

print(f"Fitted Peak Energy (Compton Scattered Peak): {fitted_peak_energy} keV")
print(f"Fitted Peak Value: {fitted_peak_value}")

# Calculate the expected Compton edge energy
source = 'na22-2'
theta = 30  # Scattering angle in degrees
expected_energy = compton_energy_for_source(source, theta)
print(f"Expected Compton Scattered Energy: {expected_energy:.2f} keV")

# Plot the smoothed data and the results
plt.figure(figsize=(10, 6))
plt.plot(with_al_data['Energy E / keV'], with_al_data['Events N'], label='With Aluminum Data (Smoothed)', alpha=0.5)
plt.plot(no_al_data['Energy E / keV'], no_al_data['Events N'], label='No Aluminum Data (Smoothed)', alpha=0.5)
plt.plot(subtracted_data['Energy E / keV'], subtracted_data['Events N'], label='Subtracted Data')
plt.plot(x_data, gaussian(x_data, *popt), label='Gaussian Fit', linestyle='--')

# Plot vertical lines for the fitted peak and expected energy
plt.axvline(x=fitted_peak_energy, color='r', linestyle='--', label=f'Fitted Peak at {fitted_peak_energy:.2f} keV')
plt.axvline(x=calibrate_energy(expected_energy), color='g', linestyle='--', label=f'Expected Energy at {expected_energy:.2f} keV')

plt.xlabel('Energy E / keV')
plt.ylabel('Events N')
plt.title('Smoothed Data with Subtraction and Gaussian Fit')
plt.xticks(np.arange(0, 2000, 100))
plt.legend()
plt.grid(True)
plt.show()

# Print calibrated energies
calibrated_fitted_peak_energy = calibrate_energy(fitted_peak_energy)
calibrated_expected_energy = expected_energy

print(f'Measured (Calibrated) Peak Energy: {calibrated_fitted_peak_energy:.2f} keV')
print(f'Expected (Calibrated) Compton Energy: {calibrated_expected_energy:.2f} keV')
```

# Math Stuff (Compton Scattering)
- Assume each photon undergoes one scattering event only, not true in dense/thick material
- The electron is assumed to be free and stationary (not bound to an atom). Not true for low energy photons where ($E << m_e c^2$)
- Assumed the source radiation profile is accurate. 
- Assume scattering angle is fixed and uniform for all photons. 
- Klein-Nishina Formula (Point like detector, no photon polarization)
- perfect energy calibration and linear
- no background and no detector broadening 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Constants
m_e_c2 = 511  # keV (electron rest mass energy)
calibration_factor = 7.78  # Calibration factor to adjust energy values

# Function to calculate theoretical Compton energy
def compton_energy(E, theta_deg):
    theta_rad = np.radians(theta_deg)
    return E / (1 + (E / m_e_c2) * (1 - np.cos(theta_rad)))

# Load dataset (example file: radiation_profiles.csv with 'Angle', 'Energy', 'Counts')
data = pd.read_csv("data/experiment/bin/1126-512bin-1gain/Cs137-sspht022-45.csv")

# Apply the calibration factor to the energy column
data['Energy E / keV'] *= calibration_factor

# Group data by angle
angles = data['Angle'].unique()
initial_energy = 661.7  # Example initial photon energy (keV) for Cs-137

# Initialize results
results = []

# Analyze spectrum for each angle
for angle in angles:
    spectrum = data[data['Angle'] == angle]
    energy = spectrum['Energy E / keV']
    counts = spectrum['Events N']
    
    # Identify the Compton peak (broad spectrum peak)
    peaks, _ = find_peaks(counts, height=0.1 * max(counts))  # Detect peaks
    if len(peaks) > 0:
        peak_energy = energy.iloc[peaks[0]]  # Energy of the first major peak
        theoretical_energy = compton_energy(initial_energy, angle)
        
        results.append({
            'Angle (deg)': angle,
            'Experimental E\' (keV)': peak_energy,
            'Theoretical E\' (keV)': theoretical_energy,
            'Difference (keV)': abs(peak_energy - theoretical_energy)
        })

        # Plot spectrum with identified peak
        plt.plot(energy, counts, label=f"Angle: {angle}°")
        plt.scatter([peak_energy], [counts.iloc[peaks[0]]], color='red', label=f"Peak: {peak_energy:.1f} keV")

# Display plot
plt.xlabel("Energy (keV)")
plt.ylabel("Counts")
plt.title("Radiation Spectra at Different Angles")
plt.legend()
plt.grid(True)
plt.show()
```

# Curve Fitting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

# Constants
m_e_c2 = 511  # keV (electron rest mass energy)

# Compton scattering formula
def compton_scatter_energy(E, theta_deg):
    theta_rad = np.radians(theta_deg)
    return E / (1 + (E / m_e_c2) * (1 - np.cos(theta_rad)))

# Klein-Nishina scaling
def klein_nishina_scale(E, E_prime, theta_deg):
    theta_rad = np.radians(theta_deg)
    return (E_prime / E) ** 2 * (E / E_prime + E_prime / E - np.sin(theta_rad) ** 2)

# Function to produce scattered radiation profile
def compton_scatter_profile(data, theta_deg):
    original_energy = data['Energy E / keV']
    original_counts = data['Events N']

    # Compute scattered energies
    scattered_energy = compton_scatter_energy(original_energy, theta_deg)

    # Compute intensity scaling factor
    scaling_factors = klein_nishina_scale(original_energy, scattered_energy, theta_deg)
    scattered_counts = original_counts * scaling_factors

    # Interpolate to align with the original energy axis
    interp_func = interp1d(scattered_energy, scattered_counts, bounds_error=False, fill_value=0)
    new_counts = interp_func(original_energy)

    # Create new DataFrame for the scattered profile
    scattered_profile = pd.DataFrame({
        'Energy E / keV': original_energy,
        'Scattered Counts': new_counts
    })

    return scattered_profile

# Filter data for energy range between 0 and 100 keV
data = data[(data['Energy E / keV'] >= 0) & (data['Energy E / keV'] <= 100)]

# Scatter profile at 45 degrees
theta = 45  # Scattering angle in degrees
scattered_profile = compton_scatter_profile(data, theta)

# Plot the original and scattered profiles
plt.plot(data['Energy E / keV'], data['Events N'], label="Original Profile")
plt.plot(scattered_profile['Energy E / keV'], scattered_profile['Scattered Counts'], label=f"Expected Scattered Profile (θ = {theta}°)")
plt.xlabel("Energy (keV)")
plt.ylabel("Counts")
plt.title("Compton Scattered Radiation Profile")
plt.legend()
plt.grid(True)
plt.show()

# Similarity metrics
def calculate_similarity(measured, expected):
    # Ensure arrays are aligned
    measured = np.array(measured)
    expected = np.array(expected)

    # RMSE
    rmse = np.sqrt(mean_squared_error(measured, expected))

    # Correlation Coefficient
    corr_coef, _ = pearsonr(measured, expected)

    # Cosine Similarity
    cosine_sim = np.dot(measured, expected) / (np.linalg.norm(measured) * np.linalg.norm(expected))

    # Offset (mean difference)
    offset = np.mean(measured - expected)

    return {
        "RMSE": rmse,
        "Correlation Coefficient": corr_coef,
        "Cosine Similarity": cosine_sim,
        "Offset": offset
    }

# Align measured and calculated profiles for comparison
measured_counts = data['Events N'].values
expected_counts = scattered_profile['Scattered Counts'].values

# Remove NaN values from both arrays
valid_indices = ~np.isnan(measured_counts) & ~np.isnan(expected_counts)
measured_counts = measured_counts[valid_indices]
expected_counts = expected_counts[valid_indices]

# Calculate similarity metrics
similarity_metrics = calculate_similarity(measured_counts, expected_counts)


# Display results
for metric, value in similarity_metrics.items():
    print(f"{metric}: {value:.4f}")
```
