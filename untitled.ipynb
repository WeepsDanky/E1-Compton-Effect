# Math Stuff (Compton Scattering)
- Assume each photon undergoes one scattering event only, not true in dense/thick material
- The electron is assumed to be free and stationary (not bound to an atom). Not true for low energy photons where ($E << m_e c^2$)
- Assumed the source radiation profile is accurate. 
- Assume scattering angle is fixed and uniform for all photons. 
- Klein-Nishina Formula (Point like detector, no photon polarization)
- perfect energy calibration and linear
- no background and no detector broadening 

```python
# Read the text file into a DataFrame, skipping the first 5 rows
df_txt = pd.read_csv('data/experiment/Cs137-sspht022-45.txt', delimiter='\t', skiprows=6)  # Adjust the delimiter as needed

# Convert columns to numeric
df_txt = df_txt.apply(pd.to_numeric, errors='coerce')

# Drop rows with all NaN values
df_txt = df_txt.dropna(how='all')
# Add columns to the DataFrame
df_txt.columns = ['Time (s)', 'Events N', 'Channel n', 'Energy E / keV', 'Rate R / 1/s']

# Save the DataFrame to a CSV file
df_txt.to_csv('data/experiment/Cs137-sspht022-45.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Constants
m_e_c2 = 511  # keV (electron rest mass energy)
calibration_factor = 7.78  # Calibration factor to adjust energy values

# Function to calculate theoretical Compton energy
def compton_energy(E, theta_deg):
    theta_rad = np.radians(theta_deg)
    return E / (1 + (E / m_e_c2) * (1 - np.cos(theta_rad)))

# Load dataset (example file: radiation_profiles.csv with 'Angle', 'Energy', 'Counts')
data = pd.read_csv("data/experiment/Cs137-sspht022-45.csv")

# Apply the calibration factor to the energy column
data['Energy E / keV'] *= calibration_factor

# Group data by angle
angles = data['Angle'].unique()
initial_energy = 661.7  # Example initial photon energy (keV) for Cs-137

# Initialize results
results = []

# Analyze spectrum for each angle
for angle in angles:
    spectrum = data[data['Angle'] == angle]
    energy = spectrum['Energy E / keV']
    counts = spectrum['Events N']
    
    # Identify the Compton peak (broad spectrum peak)
    peaks, _ = find_peaks(counts, height=0.1 * max(counts))  # Detect peaks
    if len(peaks) > 0:
        peak_energy = energy.iloc[peaks[0]]  # Energy of the first major peak
        theoretical_energy = compton_energy(initial_energy, angle)
        
        results.append({
            'Angle (deg)': angle,
            'Experimental E\' (keV)': peak_energy,
            'Theoretical E\' (keV)': theoretical_energy,
            'Difference (keV)': abs(peak_energy - theoretical_energy)
        })

        # Plot spectrum with identified peak
        plt.plot(energy, counts, label=f"Angle: {angle}°")
        plt.scatter([peak_energy], [counts.iloc[peaks[0]]], color='red', label=f"Peak: {peak_energy:.1f} keV")

# Display plot
plt.xlabel("Energy (keV)")
plt.ylabel("Counts")
plt.title("Radiation Spectra at Different Angles")
plt.legend()
plt.grid(True)
plt.show()
```

# Expected Compton Scattering Value

```python
import numpy as np

# Constants
M_E_C2 = 511  # keV (electron rest mass energy)

# Dictionary of common photon sources with their primary photon energy in keV
PHOTON_ENERGIES = {
    "cs137": 661.7,  # Cs-137 gamma-ray source
    "co60": 1173.2,  # Co-60 gamma-ray source
    "na22": 1274.5,  # Na-22 gamma-ray source
}

def compton_energy_for_source(source, angle_deg):
    """
    Calculate the expected scattered energy for a given source and angle.
    
    Parameters:
    - source (str): Name of the source (e.g., 'cs137').
    - angle_deg (float): Scattering angle in degrees.
    
    Returns:
    - float: Theoretical scattered photon energy in keV.
    """
    # Get the initial photon energy for the source
    source = source.lower()
    if source not in PHOTON_ENERGIES:
        raise ValueError(f"Source '{source}' not found in the photon energies database.")
    
    E = PHOTON_ENERGIES[source]  # Initial photon energy in keV
    theta_rad = np.radians(angle_deg)  # Convert angle to radians
    
    # Compton formula
    E_prime = E / (1 + (E / M_E_C2) * (1 - np.cos(theta_rad)))
    return E_prime

# Example usage
source_name = "cs137"
scattering_angle = 45  # in degrees
scattered_energy = compton_energy_for_source(source_name, scattering_angle)

print(f"Theoretical scattered energy for {source_name.upper()} at {scattering_angle}°: {scattered_energy:.2f} keV")
```

# Curve Fitting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

# Constants
m_e_c2 = 511  # keV (electron rest mass energy)

# Compton scattering formula
def compton_scatter_energy(E, theta_deg):
    theta_rad = np.radians(theta_deg)
    return E / (1 + (E / m_e_c2) * (1 - np.cos(theta_rad)))

# Klein-Nishina scaling
def klein_nishina_scale(E, E_prime, theta_deg):
    theta_rad = np.radians(theta_deg)
    return (E_prime / E) ** 2 * (E / E_prime + E_prime / E - np.sin(theta_rad) ** 2)

# Function to produce scattered radiation profile
def compton_scatter_profile(data, theta_deg):
    original_energy = data['Energy E / keV']
    original_counts = data['Events N']

    # Compute scattered energies
    scattered_energy = compton_scatter_energy(original_energy, theta_deg)

    # Compute intensity scaling factor
    scaling_factors = klein_nishina_scale(original_energy, scattered_energy, theta_deg)
    scattered_counts = original_counts * scaling_factors

    # Interpolate to align with the original energy axis
    interp_func = interp1d(scattered_energy, scattered_counts, bounds_error=False, fill_value=0)
    new_counts = interp_func(original_energy)

    # Create new DataFrame for the scattered profile
    scattered_profile = pd.DataFrame({
        'Energy E / keV': original_energy,
        'Scattered Counts': new_counts
    })

    return scattered_profile

# Filter data for energy range between 0 and 100 keV
data = data[(data['Energy E / keV'] >= 0) & (data['Energy E / keV'] <= 100)]

# Scatter profile at 45 degrees
theta = 45  # Scattering angle in degrees
scattered_profile = compton_scatter_profile(data, theta)

# Plot the original and scattered profiles
plt.plot(data['Energy E / keV'], data['Events N'], label="Original Profile")
plt.plot(scattered_profile['Energy E / keV'], scattered_profile['Scattered Counts'], label=f"Expected Scattered Profile (θ = {theta}°)")
plt.xlabel("Energy (keV)")
plt.ylabel("Counts")
plt.title("Compton Scattered Radiation Profile")
plt.legend()
plt.grid(True)
plt.show()

# Similarity metrics
def calculate_similarity(measured, expected):
    # Ensure arrays are aligned
    measured = np.array(measured)
    expected = np.array(expected)

    # RMSE
    rmse = np.sqrt(mean_squared_error(measured, expected))

    # Correlation Coefficient
    corr_coef, _ = pearsonr(measured, expected)

    # Cosine Similarity
    cosine_sim = np.dot(measured, expected) / (np.linalg.norm(measured) * np.linalg.norm(expected))

    # Offset (mean difference)
    offset = np.mean(measured - expected)

    return {
        "RMSE": rmse,
        "Correlation Coefficient": corr_coef,
        "Cosine Similarity": cosine_sim,
        "Offset": offset
    }

# Align measured and calculated profiles for comparison
measured_counts = data['Events N'].values
expected_counts = scattered_profile['Scattered Counts'].values

# Remove NaN values from both arrays
valid_indices = ~np.isnan(measured_counts) & ~np.isnan(expected_counts)
measured_counts = measured_counts[valid_indices]
expected_counts = expected_counts[valid_indices]

# Calculate similarity metrics
similarity_metrics = calculate_similarity(measured_counts, expected_counts)


# Display results
for metric, value in similarity_metrics.items():
    print(f"{metric}: {value:.4f}")
```
